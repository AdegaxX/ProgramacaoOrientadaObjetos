1. Faça um programa que simule um lançamento de dados. Lance o dado 100
vezes e armazene os resultados em um vetor. Depois, monte um outro vetor
contendo quantas vezes cada valor foi obtido. Imprima os dois vetores. Use a
função random.randint(1,6) para gerar números aleatórios, simulando os
lançamentos dos dados.

```python
# 1.
import random

resultados = [] # os dados serão armazenados nesta lista vazia.

# Criando a lista aleatória dos dados lançados:
for i in range(100):  # irá criar 100 execuções dos códicos.
  resul_dados = random.randint(1,6)
  resultados.append(resul_dados)  # irá adicionar na lista 'resultados' cada random executado na linha acima.

# Criando contador para os lançamentos dos valores:
cont_resultados = [0]*6 # cria um formato para a contagem dos lançamentos, 6 dados representa as faces dos dados

for resul_dados in resultados:
  cont_resultados[resul_dados - 1] += 1 # o 'resul_dados - 1' é necessário para converter os valores de '1 a 6'(usado para contar a face dos dados) para '0 a 5'(que representará os índices).

# Criando um for para mostrar os resultados:
for c in range(6):
  print('O valor de', c + 1, 'aparece', cont_resultados[c], 'vezes')

#print(resultados)
#print(cont_resultados)
```

2. Faça um programa que percorre um vetor e imprime na tela a média dos
valores do vetor e o valor mais próximo da média.

```python
# 2.
vetor = []
for i in range(5):
  vetor.append(float(input('Digite um valor para o vetor: ')))

# Calcular a média dos valores no vetor
soma = 0.0
for valor in vetor:
    soma += valor

media = soma / len(vetor)
print('\n')
# Encontrar o valor mais próximo da média
valor_proximo = None
distancia_minima = float('inf')  # Inicializa com um valor infinito

for valor in vetor:
    distancia = abs(valor - media)  # 'abs' significa 'valor absoluto', ou seja, o modulo do valor para não trabalhar com números negativos
    if distancia < distancia_minima:
        distancia_minima = distancia
        valor_proximo = valor

# Imprimir a média e o valor mais próximo
print("Vetor:", vetor)
print("Média dos valores:", media)
print("Valor mais próximo da média:", valor_proximo)
```

3. Faça um programa que percorre duas listas e intercala os elementos de
ambas, formando uma terceira lista. A terceira lista deve começar pelo
primeiro elemento da lista menor.

```python
# Criando as listas maior e menor e a lista vazia para armazenar a concatenação delas;
lista_menor = [1,3,5,7]
lista_maior = [2,4,6,8,10,12,14]
lista_concat = []

for i in range(len(lista_menor)): # concatena todos os 4 elementos da lista menor com os 4 elementos da lista menor
    lista_concat.append(lista_menor[i])
    lista_concat.append(lista_maior[i])

# Adicionar os elementos restantes da lista maior
for i in range(len(lista_menor), len(lista_maior)): # é como se fosse 'range(4, 7)'
    lista_concat.append(lista_maior[i])

print(lista_concat)
```
